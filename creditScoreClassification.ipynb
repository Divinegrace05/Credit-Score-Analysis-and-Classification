### Data Preprocessing

```python
# Drop Unnecessary Columns
df.drop(columns=['ID', 'Customer_ID', 'SSN', 'Name'], inplace=True)
```

Checkng for inconsistencies

```python
for col in df.columns:
    print(f"Column: {col}")
    print(df[col].value_counts(dropna=False))  # Include NaN counts
    print("-" * 50)  # Separator for readability
```

A valid human age range (esp. one that is eligible for loans) should be 18 to 100.

```python
# Fix Age Column (Remove Invalid Values)
df = df[df['Age'].astype(str).str.isnumeric()]  # Keep only numeric values
df['Age'] = df['Age'].astype(int)  # Convert to integer
df = df[df['Age'] > 0]  # Remove negative ages
df = df[(df['Age'] >= 18) & (df['Age'] <= 100)]  

# Convert Categorical Columns
df['Credit_Mix'] = df['Credit_Mix'].replace('_', np.nan)

df['Occupation'] = df['Occupation'].replace("_______", np.nan)  # Convert to NaN
df['Occupation'].fillna(df['Occupation'].mode()[0], inplace=True)  # Impute with mode

df['Type_of_Loan'] = df['Type_of_Loan'].str.split(',').str[0]  # Extract only the first loan type
df['Type_of_Loan'] = df['Type_of_Loan'].str.strip()  # Remove extra spaces
df['Type_of_Loan'].fillna("Unknown", inplace=True)  # Fill missing values

df['Payment_of_Min_Amount'] = df['Payment_of_Min_Amount'].replace("NM", df['Payment_of_Min_Amount'].mode()[0])
df['Payment_Behaviour'] = df['Payment_Behaviour'].replace("!@9#%8", df['Payment_Behaviour'].mode()[0])

df['Delay_from_due_date'] = df['Delay_from_due_date'].apply(lambda x: max(x, 0))
```

Encoding Categorical Variables

Handling Missing Values

```python
# Check Missing Value Percentage
missing_values = (df.isnull().sum() / len(df)) * 100
missing_values[missing_values > 0].sort_values(ascending=False)
```

All columns have moderate amount of missing values (10-40%), we can impute values.

```python
# Convert Credit_History_Age to Numeric Format
import re

def convert_credit_history(age_str):
    if pd.isnull(age_str):
        return np.nan
    match = re.search(r'(\d+)\s+Years?\s+and\s+(\d+)\s+Months?', age_str)
    if match:
        years = int(match.group(1))
        months = int(match.group(2))
        return years * 12 + months
    return np.nan

df['Credit_History_Age'] = df['Credit_History_Age'].apply(convert_credit_history)

numeric_cols = ['Monthly_Inhand_Salary', 'Num_of_Delayed_Payment', 'Credit_History_Age',
                'Amount_invested_monthly', 'Annual_Income', 'Num_of_Loan', 'Changed_Credit_Limit',
                'Num_Credit_Inquiries', 'Monthly_Balance', 'Outstanding_Debt']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert non-numeric values to NaN

# Fill numerical columns with median
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Fill categorical column 'Type_of_Loan' with "Unknown"
df['Type_of_Loan'].fillna('Unknown', inplace=True)
df['Credit_Mix'].fillna(df['Credit_Mix'].mode()[0], inplace=True)
df['Type_of_Loan'].fillna("Unknown", inplace=True)
```

```python
df.loc[df['Changed_Credit_Limit'] < 0, 'Changed_Credit_Limit'] = np.nan  # Replace negatives with NaN

# Fill NaN with the median value (more robust than mean)
df['Changed_Credit_Limit'].fillna(df['Changed_Credit_Limit'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encoding target variable
le = LabelEncoder()
df['Credit_Score'] = le.fit_transform(df['Credit_Score'])

df['Payment_Behaviour'] = df['Payment_Behaviour'].str.split('_').str[0]  # Keep only "High" or "Low"
df['Payment_Behaviour'] = df['Payment_Behaviour'].replace({'High': 1, 'Low': 0})
```

```python
# First, split and explode the loan types into separate rows
df_exploded = df.assign(Type_of_Loan=df['Type_of_Loan'].str.split(',')).explode('Type_of_Loan')

# One-hot encode the loan types
loan_dummies = pd.get_dummies(df_exploded['Type_of_Loan'], prefix='Loan')

# Merge back to the original dataset
df = df.drop(columns=['Type_of_Loan']).join(loan_dummies.groupby(df_exploded.index).max())
```
